## Toolkits 

A toolkit is just a collection of related tools that serve a common purpose packaged together for convinience and reusability.

In langchain:
  A toolkit might be : GoogleDriveToolKit
  And it can contain the following Tools

  1. GoogleDriveCreateFileTool --> upload a file
  2. GoogleDriveSearchTool  ---> search a file by name/content
  3. GoogleDriveReadFileTool   ---> read contents of file


```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
