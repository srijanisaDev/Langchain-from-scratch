## Built-in-Tools

A built-in tool is a tool that langchain already provides for you - it's pre-built , production-ready and requires minimal or no setup.

You dont have to write the function logic -- just import and use

```python
!pip install langchain langchain-core langchain-community pydantic duckduckgo-search
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
```

## Built in tool - DuckDuckgosearch

```python
search_tool = DuckDuckGoSearchRun()
```

```python
%pip install -U ddgs
```

```python
results = search_tool.invoke('news on america iran conflict')
```

```python
print(results)
```
