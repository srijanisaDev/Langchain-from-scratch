## Custom Tools 

A custom tool is a tool that you define yourself.

use them when :
 you need to call your own APIs ,
 you want to encapsulate bussiness logic , 
 you want the LLM to interact with your database , product or app.

```python
from langchain_community.tools import tool
```

```python
# step 1 create a function

def multiply(a,b):
    """Multiply two numbers"""
    return a*b
```

```python
# step 2 type hint

def multiply(a: int,b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
# step 3 - add tool decortator

@tool
def multiply(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({"a" : 7 , "b" : 8})
```

```python
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema)
```

## Two more ways to create tools other than @tool

## Structured Tool

A structured tool in langchain is a special type of tool where the input to the tool follows a structured schema , typically defined using a pydantic model.

```python
from langchain_community.tools import StructuredTool
```

```python
from pydantic import BaseModel , Field , json_schema
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True , description="The first number to add")
    b: int = Field(required=True , description="The second number to add")
```

```python
def multiply_func(a: int , b: int):
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3 , 'b' : 12})
```

```python
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
```

## Base Tool Class

is the abstract base class for all tools in langchain . It defines the cre structure and interface that any tool must follow , whether it's a simple one-linear or a fully customized function.

All other tool types like @tool , StructuredTool are built on top of BaseTool

```python
from langchain_community.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':11})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```
