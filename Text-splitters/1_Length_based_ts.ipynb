## Length based Text splitting

One of the most simplest and fastest method of chinking a large text . In this we pre-determine the chunk size and the splitter does its job .

Onne of the disadvantages of this is that it splits the chunk in characters and it doesnot follow a linguistic structure ....if the chunk1 ends in between the word "determine" then chunk1--->.........deter , chunk2--->mine........and the semantic meaning of the words also changes.

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text = """
The old lighthouse at the edge of Morrow's Cliff had not been lit in over forty years, yet sailors still claimed to see its beam cutting through the fog on stormy nights. Some attributed it to reflections off the jagged rocks below, while others — the more superstitious kind — whispered about the keeper who had refused to abandon his post even as the sea swallowed the lower steps whole. The town never bothered to investigate; some mysteries, the locals agreed, were better left to the wind and the salt and the dark water below.


"""
```

```python
splitter = CharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 0,
    separator=''
)
```

One of the benefits of using chunk overlap is that it helps retain the semantic meaning of a word when we are abruptly splitting text . when set to zero means no overlapping happens .  visit ---> https://chunkviz.up.railway.app/ for visualization.

```python
result = splitter.split_text(text)
```

```python
print(result)
```
