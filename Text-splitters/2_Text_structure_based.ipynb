```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = """
The old lighthouse at the edge of Morrow's Cliff had not been lit in over forty years, yet sailors still claimed to see its beam cutting through the fog on stormy nights. Some attributed it to reflections off the jagged rocks below, while others — the more superstitious kind — whispered about the keeper who had refused to abandon his post even as the sea swallowed the lower steps whole. The town never bothered to investigate; some mysteries, the locals agreed, were better left to the wind and the salt and the dark water below.


"""
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 0,
)
```

```python
chunks = splitter.split_text(text)
```

```python
print(len(chunks))
```

```python
print(chunks)
```
