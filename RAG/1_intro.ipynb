## RAG
It is a technique that combines information with language generation, where a model retrives relevent documents from a knowledge base and then uses them as context to generate accurate and grounded responses

Benefits of using RAG

1. Use of up to date information
2. Better privacy
3. No limit of document size

## Components of RAG

1. Document Loaders
2. Text Splitters
3. Vector Databases
4. Retrievers
