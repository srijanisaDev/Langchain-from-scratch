```python
from langchain_community.document_loaders import DirectoryLoader , PyPDFLoader
```

```python
loader = DirectoryLoader(
    path = 'books',
    glob = '*.pdf',
    loader_cls = PyPDFLoader
)
```

```python
docs = loader.load()
```

```python
print(len(docs))  ## the output is 346 because it is the number of pages in all the pdfs combined 
```
