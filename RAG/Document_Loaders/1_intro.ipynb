## Document Loaders 

Document loaders are components in langchain used to load data from various sources into a standardized format
(usually as Document objects ) , which can then be used for chunking , embedding , retrieval , and generation.

## Text Loader

is a simple and commonly used document loader in langchain that reads plain text .txt files and converts them into langcahin document objects.

## Usecase
ideal for loading chat logs , scraped text,transcripts , code snippets , or any plain text data into a langchain pipeline.

## Limitation
works only with .txt files

## PyPPDF loader

is a doc loader in langchain used to load content from pdf files and convert each page into a document object.

## Usecases:
1. simple cleaned pdfs ---> PyPDF Loader
2. PDFs with tables and cols ---> PDFlumberLoader
3. Scanned/ image PDFs ---> UnstructurePDFLoader or AmazonTextractPDFLoader
4. Need layout and image data ---> PyMuPDFLoader
5. Want best Structure Extraction ---> UnstructurePDF Loader


## Limitations:
  it uses PyPDF library under the hood -- not great with scanned PDFs and complex layouts.

## Directory Loader

is a doc loader that lets you load multiple documents from a directory (folder) of files.
