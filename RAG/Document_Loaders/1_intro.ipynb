## Document Loaders 

Document loaders are components in langchain used to load data from various sources into a standardized format
(usually as Document objects ) , which can then be used for chunking , embedding , retrieval , and generation.

## Text Loader

is a simple and commonly used document loader in langchain that reads plain text .txt files and converts them into langcahin document objects.

## Usecase
ideal for loading chat logs , scraped text,transcripts , code snippets , or any plain text data into a langchain pipeline.

## Limitation
works only with .txt files
