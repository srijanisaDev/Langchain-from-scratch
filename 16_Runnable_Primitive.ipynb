## Runnable Primitives
These are fundamental building blocks for structuring execution logic in AI Workflows

Types od Runnable Primitives

## 1. Runnable Sequence
Runnable Sequence is a sequential chain of runnables in Lnagchain that executes each step one after another , passing the output of one step as the input of the next.

It is useful when you need to compare multiple runnables together in a structured workflow.
