## Structured ouput 

# In langchain , structured output refers to the practices of having language models return rsponses in a well-defined data format (for example - JSON) , rather than free-from text . This  makes the output model easier to parse and work with programmatically.

## Ways to get structured output

There are LLM's which can provide both "with structured output" and "not structured output" for those who cannot provide with structured outputs we use output parsers for them
